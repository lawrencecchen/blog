Raspberry Pi Remote Control Car
> Making a remote controlled robot car with a Raspberry Pi and the EECS 16b lab car.

- toc: false
- layout: post
- categories: [projects]

Berkeley's EECS 16b final lab is a voice controlled car. The courses uses a TI Launchpad (which is really outdated and slow), and only roughly four commands can fit in memory (even after computing the minimum norm to pre-computed priniciple component vectors). I decided to replace the entire system with a remote controller app (full-screen website) and a Raspberry Pi.
